```python
# URL of page to be scraped
base_url = 'https://www.boxofficemojo.com'
url_l1 = 'https://www.boxofficemojo.com/year/?ref_=bo_nb_di_secondarytab'
```

```python
# Retrieve page with the requests module
response_l1 = requests.get(url_l1)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup_l1 = BeautifulSoup(response_l1.text, 'html.parser')
```

```python
years_td = soup_l1.find_all('td',class_='mojo-field-type-year')
```

```python
no_of_years = 21
# no_of_years = 2
movies_det_list = []
```

```python
start_time = time.time()
```

```python
for year_td in years_td[:no_of_years]:
    year = year_td.text
    year_link = year_td.find('a')['href']
    
    url_l2 = base_url + year_link
    response_l2 = requests.get(url_l2)
    soup_l2 = BeautifulSoup(response_l2.text, 'html.parser')
    movies_td = soup_l2.find_all('td',class_='mojo-field-type-release')
    
    for movie_td in movies_td:
        movie_dict={}
        movie=dom_gross=int_gross=world_gross=distributor=budget=running_time=theatres=''
        genres=[]
        
        movie = movie_td.text
        movie_link = movie_td.find('a')['href']
        
        url_l3 = base_url + movie_link
        response_l3 = requests.get(url_l3)
        soup_l3 = BeautifulSoup(response_l3.text, 'html.parser')
        
        try:
            dom_gross = (soup_l3.find('h2',text=re.compile('Grosses'))
                         .find_next_siblings()[0].find('span',class_='money').text
                        )
        except:
            pass

        try:
            int_gross = (soup_l3.find('h2',text=re.compile('Grosses'))
                         .find_next_siblings()[1].find('span',class_='money').text
                        )
        except:
            pass

        try:
            world_gross = (soup_l3.find('h2',text=re.compile('Grosses'))
                           .find_next_siblings()[2].find('span',class_='money').text
                          )
        except:
            pass

        try:
            distributor = soup_l3.find('span',text='Distributor').find_next_sibling().contents[0]
        except:
            pass

        try:
            budget = soup_l3.find('span',text='Budget').find_next_sibling().find('span').text
        except:
            pass

        try:
            running_time = soup_l3.find('span',text='Running Time').find_next_sibling().text
        except:
            pass

        try:
            genres = soup_l3.find('span',text='Genres').find_next_sibling().text.split()
        except:
            pass

        try:
            theatres = soup_l3.find('span',text='Widest Release').find_next_sibling().text.split()[0]
        except:
            pass

        if not any(mov.get('movie','XXXXXXX') == movie for mov in movies_det_list):

            movie_dict['movie'] = movie
            movie_dict['domestic_gross'] = dom_gross
            movie_dict['international_gross'] = int_gross
            movie_dict['world_gross'] = world_gross
            movie_dict['distributor'] = distributor
            movie_dict['budget'] = budget
            movie_dict['running_time'] = running_time
            movie_dict['genres'] = genres
            movie_dict['theatres'] = theatres
            movie_dict['year_released'] = year

            movies_det_list.append(movie_dict)
```
